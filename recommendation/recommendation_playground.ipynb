```python
user_playlist = [
    {"artist_name": "Dream Theater", "album_name": "Train of Thought", "track_name": "In The Name Of God"},
    {"artist_name": "Dream Theater", "album_name": "Train of Thought", "track_name": "As I Am"},
    {"artist_name": "Dream Theater", "album_name": "Train of Thought", "track_name": "Stream of Consciousness"},
    {"artist_name": "Nightwish", "album_name": "Imaginaerum", "track_name": "Sundown"},
    {"artist_name": "Nightwish", "album_name": "Imaginaerum", "track_name": "Dare to Enter"},
    {"artist_name": "Gojira", "album_name": "Magma", "track_name": "Silvera"},
    {"artist_name": "Gojira", "album_name": "Magma", "track_name": "Stranded"},
    {"artist_name": "Gojira", "album_name": "L'Enfant Sauvage", "track_name": "L'Enfant Sauvage"},
    {"artist_name": "Gojira", "album_name": "L'Enfant Sauvage", "track_name": "Explosia"}
]
```

```python
artist_name_set = set([user_track['artist_name'] for user_track in user_playlist])
album_name_set = set([re.sub('\W+','', user_track['album_name'].lower()) for user_track in user_playlist])
track_name_set = set([re.sub('\W+','', user_track['track_name'].lower()) for user_track in user_playlist])

artist_filter = "|".join(artist_name_set)
album_filter = "|".join(album_name_set)
track_filter = "|".join(track_name_set)

print(artist_filter)
print(album_filter)
print(track_filter)

df_filter = lambda x: re.sub('\W+','', x.lower())

user_playlist_df = tracks_df[tracks_df['artist_name'].str.contains(artist_filter, regex=True)]
user_playlist_df = user_playlist_df[user_playlist_df['album_name'].apply(df_filter).str.contains(album_filter)]
user_playlist_df = user_playlist_df[user_playlist_df['track_name'].apply(df_filter).str.contains(track_filter)]

print(len(user_playlist_df.index))
display(user_playlist_df)
```

```python
user_playlist_genre_df = user_playlist_df[['genre', 'r', 'g', 'b', 'font_size']].reset_index(drop=True)
display(user_playlist_genre_df)
playlist_length = len(user_playlist_genre_df.index)
genre_set = set(user_playlist_df.genre.values)
```

```python
user_playlist_genre_df = user_playlist_genre_df.groupby(['genre']).agg({
    'r': 'mean', 
    'g': 'mean', 
    'b': 'mean', 
    'font_size': ['min', 'max', 'mean', 'median', 'std', 'var'], 
    'genre': 'count'
})
display(user_playlist_genre_df)
```

```python
genres_list = [v for v in user_playlist_genre_df.to_dict().values()]   
print(genres_list)
rs = genres_list[0]
gs = genres_list[1]
bs = genres_list[2]
pop_mins = genres_list[3]
pop_maxs = genres_list[4]
print(rs)
print(gs)
print(bs)
print(pop_mins['progressive metal'])
print(pop_maxs['progressive metal'])
closest_genres_df = None

for i in range(0, len(rs.values())):
    r = [v for v in rs.values()][i]
    g = [v for v in gs.values()][i]
    b = [v for v in bs.values()][i]

    rgb_epsilon = 5.0
    closest_genres_df = close_genres_df = tracks_df.groupby(['genre']).agg({
        'r': 'mean', 
        'g': 'mean', 
        'b': 'mean'
    })
    selector = (closest_genres_df['r'] >= r - rgb_epsilon) & (closest_genres_df['r'] <= r + rgb_epsilon) \
    & (closest_genres_df['g'] >= g - rgb_epsilon) & (closest_genres_df['g'] <= g + rgb_epsilon) \
    & (closest_genres_df['b'] >= b - rgb_epsilon) & (closest_genres_df['b'] <= b + rgb_epsilon)

    closest_genres_df = closest_genres_df[selector]
    display(closest_genres_df.head())
```

```python
closest_genres_df = closest_genres_df.sample(5)
display(closest_genres_df)
```

```python
popularity_epsilon = 15.0
recommended_genres_filter = '|'.join(closest_genres_df.index.values.tolist())
display(recommended_genres_filter)

recommended_genres_tracks_df = tracks_df[
    (tracks_df['genre'].str.contains(recommended_genres_filter)) \
    & (tracks_df['font_size'] >= pop_mins['progressive metal'] - popularity_epsilon) \
    & (tracks_df['font_size'] <= pop_maxs['progressive metal'] + popularity_epsilon)
]
display(recommended_genres_tracks_df.sample(20))
```
