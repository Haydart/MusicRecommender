```python
artists_df = pd.read_csv("./output/every_noise_artists_output.csv")
print(artists_df.head())
print(len(artists_df))
artists_df = artists_df.drop_duplicates(subset=['spotify_artist_id'], keep=False)
print(len(artists_df))
```
